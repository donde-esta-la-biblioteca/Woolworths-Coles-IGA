```python
wow = pd.read_csv('2. Modified Data/Modified Woolworths (modded).csv').fillna('')
```

```python
wow.head()
```

```python
iga = iga_match_wow[['Brand_Product_Size', 'IGA Price', 'IGA Category']]
wow = wow[['Brand_Product_Size', 'WOW Price', 'WOW Category', 'WOW Specials']]
```

```python
iga.head()
```

```python
wow.head()
```

```python
iga_v_wow = pd.merge(iga, wow, on='Brand_Product_Size', how='inner')
```

```python
len(iga_v_wow)
```

```python
iga_v_wow.drop_duplicates(subset=['Brand_Product_Size', 'IGA Price', 'WOW Price'], keep=False, inplace=True)
```

```python
iga_v_wow = iga_v_wow[['Brand_Product_Size', 'IGA Price', 'WOW Price', 'IGA Category', 'WOW Category', 'WOW Specials']]
iga_v_wow
```

```python
iga_v_wow.to_csv('5. IGA vs. Woolworths/iga_v_wow.csv', index=False, encoding='utf-8-sig')
```

```python
def cheapest_seller(IGA, WOW):
    if IGA > WOW:
        return 'WOW'
    elif IGA < WOW:
        return 'IGA'
    else:
        return 'Same Price'
```

```python
iga_v_wow['Cheapest Seller'] = iga_v_wow.apply(lambda x: cheapest_seller(x['IGA Price'], x['WOW Price']), axis=1)
```

```python
iga_v_wow
```

```python
iga_v_wow['Cheapest Seller'].value_counts()
```

```python
cheapest_at_iga = iga_v_wow[iga_v_wow['Cheapest Seller'] == 'IGA']
cheapest_at_iga
```

```python
cheapest_at_iga['WOW Specials'].value_counts()
```
